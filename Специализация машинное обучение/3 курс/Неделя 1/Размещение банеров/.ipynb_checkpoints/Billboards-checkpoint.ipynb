# Prepare data

```python
with open("umn_foursquare_datasets/checkins.dat", "r") as fout:
    info = fout.read()
dataset = info.split("\n") 
```

```python
print(dataset[0])
```

```python
data = dataset[2:-3]
for i, line in enumerate(data):
    data[i] = line.split("|")
    for j, word in enumerate(data[i][:-1]):
        data[i][j] = word.replace(" ", "")
len(data)
```

```python
data_for_save = []
data_for_save.append(dataset[0].split("|"))

for j, word in enumerate(data_for_save[0]):
    data_for_save[0][j] = word.replace(" ", "")
data_for_save[0] = ",".join(data_for_save[0])

for line in data:
    if line[3] != "" and line[4] != "":
        data_for_save.append(",".join(line))

len(data_for_save)
```

```python
data_for_save[1]
```

```python
with open("Clear_data.csv", "w") as fout:
    for line in data_for_save:
        fout.write(line + "\n")
```

# Сlusterization

```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```python
data = pd.read_csv("Clear_data.csv")
data.info()
```

```python
ms = MeanShift(bandwidth=0.1)
```

```python
ms.fit(data[data.columns[3:5]].iloc[:100000])
```

```python
a = ms.predict(data[data.columns[3:5]])
```

```python
data["cluster"] = a
```

```python
clusters = np.array(list(set(a)))

d = np.zeros(len(clusters))

for item in clusters:
    d[item] = len(np.where(a == item)[0])
```

```python
need_clusters = clusters[d > 15]
```

```python
result_place = []
for cl in need_clusters:
    result_place.append(
    ",".join([str(data[data.cluster == cl].latitude.mean()), str(data[data.cluster == cl].longitude.mean())])
    )
```

```python
with open("result_places.txt", "w") as fout:
    fout.write("\n".join(result_place))
```

```python
longitudes = np.zeros(len(need_clusters))
latitudes = np.zeros(len(need_clusters))

for i, cl in enumerate(need_clusters):
    
    latitudes[i] = data[data.cluster == cl].latitude.mean()
    longitudes[i] = data[data.cluster == cl].longitude.mean()
```

```python
city = ["Los Angeles", "Miami" , "London", "Amsterdam", "Beijing", "Sydney"]

dictionary = {
    "Los Angeles" : [33.751277, -118.188740],
    "Miami" : [25.867736, -80.324116] ,
    "London" : [51.503016, -0.075479 ], 
    "Amsterdam" : [52.378894, 4.885084], 
    "Beijing": [39.366487, 117.036146],
    "Sydney": [-33.868457, 151.205134]
}
```

```python
places = pd.DataFrame({
    "clusters" : need_clusters,
    "latitudes" : latitudes,
    "longitudes" : longitudes,
    "distances" : np.zeros(len(need_clusters)),
    "city" : ["Los Angeles" for i in range(len(need_clusters))]
})
```

```python
def count_distance(i, city):
    return np.sqrt((places.latitudes[i] - dictionary[city][0])**2 + (places.longitudes[i] - dictionary[city][1])**2)
```

```python
distances = np.zeros(len(need_clusters))
cityes = ["Los Angeles" for i in range(len(need_clusters))] 
for i in range(len(places.clusters)):
    dis = np.zeros(len(city))
    for j in range(len(city)):
        dis[j] = count_distance(i, city[j])
    idx = np.where(dis == dis.min())[0][0]
    distances[i] = dis[idx]
    cityes[i] = city[idx]
    
```

```python
places["distances"] = distances
places["city"] = cityes 
```

```python
places.sort_values("distances")
```
