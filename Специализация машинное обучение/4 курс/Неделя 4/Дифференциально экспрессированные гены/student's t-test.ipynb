```python
import pandas as pd
import numpy as np
from scipy.stats import t
import statsmodels.stats.multitest as smm
```

```python
data = pd.read_csv("gene_high_throughput_sequencing.csv")
```

# Часть 1 

```python
mean = data.groupby(by = "Diagnosis").mean()
std = data.groupby(by = "Diagnosis").std(ddof = 1)
cnt = data.drop(columns = "Patient_id").groupby(by = "Diagnosis").count()
```

```python
mean
```

```python
std
```

```python
cnt
```

```python
def get_p_val(mean1, mean2, std1, std2, n1, n2):
    ts = (mean1 - mean2)/np.sqrt(std1**2/n1+std2**2/n2)
    nus = (std1**2/n1+std2**2/n2)**2/(std1**4/(n1**2*(n1-1))+std2**4/(n2**2*(n2-1)))
    ps = 2*(1- t.cdf(abs(ts), nus))
    return ps
```

```python
p_vals1= get_p_val(mean.loc["normal"].values, mean.loc["early neoplasia"].values, 
          std.loc["normal"].values, std.loc["early neoplasia"].values, 
         cnt.loc["normal"].values, cnt.loc["early neoplasia"].values)
```

```python
with open("ans1.txt", "w") as f:
    f.write(str(len(p_vals1[p_vals1 < 0.05])))
```

```python
p_vals2= get_p_val(mean.loc["early neoplasia"].values, mean.loc["cancer"].values, 
          std.loc["early neoplasia"].values, std.loc["cancer"].values, 
         cnt.loc["early neoplasia"].values, cnt.loc["cancer"].values)
```

```python
with open("ans2.txt", "w") as f:
    f.write(str(len(p_vals2[p_vals2 < 0.05])))
```

# Часть 2

```python
import statsmodels.stats.multitest as smm
```

```python
def F(C, T):
    return (T>C)*(T/C) + (T <= C)*(-C/T)
```

```python
ans1, p_vals1_cor, _ , _ = smm.multipletests(p_vals1, alpha = 0.025, method = "holm")
```

```python
Fs1 = F(mean.loc["normal"].values[ans1], mean.loc["early neoplasia"].values[ans1])
```

```python
with open("ans3.txt", "w") as f:
    f.write(str(len(Fs1[Fs1 > 1.5])))
```

```python
ans2, p_vals2_cor, _ , _ = smm.multipletests(p_vals2, alpha = 0.025, method = "holm")
Fs2 = F(mean.loc["early neoplasia"].values[ans2], mean.loc["cancer"].values[ans2])
```

```python
with open("ans4.txt", "w") as f:
    f.write(str(len(Fs2[abs(Fs2) > 1.5])))
```

```python
ans3, p_vals1_cor, _ , _ = smm.multipletests(p_vals1, alpha = 0.025, method = "fdr_bh")
Fs3 = F(mean.loc["normal"].values[ans3], mean.loc["early neoplasia"].values[ans3])

with open("ans5.txt", "w") as f:
    f.write(str(len(Fs3[abs(Fs3) > 1.5])))
```

```python
ans4, p_vals1_cor, _ , _ = smm.multipletests(p_vals2, alpha = 0.025, method = "fdr_bh")
Fs4 = F(mean.loc["early neoplasia"].values[ans4], mean.loc["cancer"].values[ans4])

with open("ans6.txt", "w") as f:
    f.write(str(len(Fs4[abs(Fs4) > 1.5])))
```
